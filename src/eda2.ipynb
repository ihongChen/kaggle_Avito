```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_pickle('../input/feats/df_trn_feat0.pkl')

mask = pd.isnull(df_train.price)
print('price is null,  cnt:{}'.format(mask.sum()))


print('deal_prob=0,\tcnt:{}'.format(np.sum(df_train[mask].deal_probability==0)))
print('deal_prob=1,\tcnt:{}'.format(np.sum(df_train[mask].deal_probability==1)))

df_plot_tmp = df_train.deal_probability.to_frame()
df_plot_tmp.hist(bins=20)
```
